```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from pylab import rcParams
rcParams['figure.figsize'] = 12,9
```

```python
my_df = pd.read_csv('D:/Python/Python_course/Data Visualisation/datasets_363619_861634_20191226-items.csv')
my_df[['brand', 'title', 'rating', 'totalReviews', 'price']]
sns_plot = sns.pairplot(
    my_df[['rating', 'totalReviews', 'price']])
sns_plot
# На pair plot графике проводим анализ зависимости рейтинга, количества обзоров и цены мобильного телефона
# Количество обзоров больше для телефонов среднеценового диапазона. Связано скорее всего с тем, что моделей среднего класса больше
# Количество обзоров больше для телефонов со средним и ниже рейтингом. Связано с большим количеством моделей данного класса.
```

```python
sns.boxplot(x="brand", y="rating", data=my_df[['brand', 'rating']])
# Большинство моделей Xiaomi со средним уровнем рейтинга 4.5 Похоже, что лучшее соотношение цена-качество
# У Huawei очень хорошие позиции в части среднего рейтинга перед известными западными игроками
# Samsung находится не в лидерах устройств-конкурентов Apple. Есть модели, с низким рейтингом.
```

```python
my_df = pd.read_csv('D:/Python/Python_course/Data Visualisation/datasets_363619_861634_20191226-items.csv')
boxplot = my_df.boxplot(by='brand', column=['price', 'totalReviews'], rot=45)
# Количество обзоров телефонов huawei не много по отношению к конкурентам, однако, на рынке данные аппараты известны своим качеством.
# стоимость телефонов OnePlus высока при достаточно низкой активности в медиа. Не хватает информации по статистике продаж для того,чтобы составить полноценную картину.
```

```python
my_df = pd.read_csv('D:/Python/Python_course/Data Visualisation/datasets_363619_861634_20191226-items.csv')
huawei_apple = my_df[(my_df['brand'] == 'HUAWEI') | (my_df['brand'] == 'Apple')]
sns.lmplot(x="price", y="rating", hue="brand", data=huawei_apple)
# На гафике отображена линейная регрессия зависимости рейтинга от цены телефонных аппаратов на базе производителя Huawei и Apple.
# Можно сделать вывод, что с ростом цены, рейтинг моделей Huawei увеличивается. Т.е. наиболее успешные модели телефонов находятся в премиум сегменте.
# У Apple в среднем получается линейная зависимость, т.е. рейтинг телефонов не зависит от цены. Т.о. можно сделать вывод, что все модели телефонов Apple пользуются популярностью.
```

```python
my_df = pd.read_csv('D:/Python/Python_course/Data Visualisation/datasets_363619_861634_20191226-items.csv')
x= my_df[['brand', 'rating']].groupby('brand').median()
px.defaults.width = 600
px.defaults.height = 600
fig = px.imshow(x)
fig.show()
# Среднее значение рейтинга телефонных аппаратов
```
