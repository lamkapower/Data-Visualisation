```python
my_df = pd.read_csv('D:/Python/Python_course/Data Visualisation/datasets_363619_861634_20191226-items.csv')
boxplot = my_df.boxplot(by='brand', column=['price', 'totalReviews'], rot=45)
#    huawei      , ,       .
#   OnePlus       .        ,   .
```

```python
my_df = pd.read_csv('D:/Python/Python_course/Data Visualisation/datasets_363619_861634_20191226-items.csv')
huawei_apple = my_df[(my_df['brand'] == 'HUAWEI') | (my_df['brand'] == 'Apple')]
sns.lmplot(x="price", y="rating", hue="brand", data=huawei_apple)
#               Huawei  Apple.
#   ,    ,   Huawei . ..        .
#  Apple     , ..      . ..   ,     Apple  .
```
